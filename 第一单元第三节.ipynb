### 1.6.1 任务一：利用Pandas对示例数据进行排序，要求升序

```python
sp = pd.DataFrame(np.random.randn(3,3),
index = list('123'),
columns = list('abc'))
sp
```

1.让行索引升序排序  
sort_index()

```python
sp.sort_index()
```

2.让列索引升序排序

```python
sp.sort_index(axis=1)
```

3.让列索引降序排序

```python
sp.sort_index(ascending=False, axis=1)
```

4.让任选两列数据同时降序排序

```python
sp.sort_values(['a', 'b'], ascending=False)
```
