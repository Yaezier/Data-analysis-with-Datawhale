### 1.1.3 任务三：每1000行为一个数据模块，逐块读取  
【思考】什么是逐块读取？为什么要逐块读取呢？  
回答：把数据分成一小块一小块的chunker，方便查看

查看数据块类型

```python
chunker = pd.read_csv('train.csv', chunksize = 1000)
print(type(chunker))
print(chunker)
```

返回的对象是TextFileReader，并且发现打印不出来，只能采用for循环一条一条看

```python
for i in chunker:
    print(i)
```

### 1.1.4 任务四：将表头改成中文，索引改为乘客ID [对于某些英文资料，我们可以通过翻译来更直观的熟悉我们的数据]  
【思考】所谓将表头改为中文其中一个思路是：将英文列名表头替换成中文。还有其他的方法吗？  
回答：目前有两种方法，一种是先把表打印出来，而后修改第一行的列名；第二种则是在读取时直接修改列名。分别操作一下。（注意这个列名要以列表的形式展现，目前没有快速复制的方法，只能一条一条复制粘贴。）

```python
df = pd.read_csv('train.csv')
df.columns=['乘客ID', '是否幸存', '乘客等级(1/2/3等舱位)', '乘客姓名', '性别', '年龄', '堂兄弟/妹个数', '父母与小孩个数', '船票信息', '票价', '客舱', '登船港口']
df.head()
```

```python
df = pd.read_csv('train.csv',names = ['乘客ID', '是否幸存', '乘客等级(1/2/3等舱位)', '乘客姓名', '性别', '年龄', '堂兄弟/妹个数', '父母与小孩个数', '船票信息', '票价', '客舱', '登船港口'], header=0)
df.head()
```

注意此处header=0，如果不加这个python就会把原来的英文列名变成内容的第一行，然后列名再改成中文的
